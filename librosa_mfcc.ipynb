# MFCC
* mfcc([y, sr, S, n_mfcc, dct_type, norm, lifter])
* librosa.feature.mfcc(y=None, sr=22050, S=None, n_mfcc=20, dct_type=2, norm='ortho', lifter=0, **kwargs)
* MFCC : Mel Spectrum(멜 스펙트럼)에서 Cepstral(켑스트럴) 분석을 통해 추출된 값


    y : np.ndarray [shape=(n,)] or None | 소리가 떨리는 세기(진폭)를 시간 순서대로 나열한 것
    sr : number > 0 [scalar] | 샘플링레이트,  1초당 샘플의 개수, 단위 Hz 또는 kHz
    S : np.ndarray [shape=(d, t)] or None | log-power Mel spectrogram (뭔지 모르겠어요..)
    n_mfcc : int > 0 [scalar] | MFCC 아웃풋 개수
    dct_type : {1, 2, 3} | Discrete cosine transform (DCT) 이산 코사인 변환
    norm : None or ‘ortho’ | 정규화, dct_type이 2 혹은 3일 때만 정규화를 지원한다.
    lifter : number >= 0 | 숫자가 클수록 선형에 가까워진다.
    kwargs : 추가적인 arguments

---------------------------

```python
import librosa
y1, sr1 = librosa.load('C:\\nmb\\nada\\testvoice_minsun2.wav')
y2, sr2 = librosa.load('C:\\nmb\\nada\\testvoice_minsun2.wav')
y3, sr3 = librosa.load('C:\\nmb\\nada\\testvoice_minsun2.wav')
```

```python
# Generate mfccs from a time series
mfccs1 = librosa.feature.mfcc(y=y1, sr=sr1)
print("y : ", y1)
print("sr : ", sr1)
print("y길이 : ", len(y1))
print("Sampling rate (KHz) : %d" % sr1)                  
print("Audio length (seconds) : %.2f" % (len(y1) / sr1)) 
```

```python
# Using a different hop length and HTK-style Mel frequencies
mfccs2 = librosa.feature.mfcc(y=y2, sr=sr2, hop_length=1024, htk=True)
print("y : ", y2)
print("sr : ", sr2)
print("y길이 : ", len(y2))
print("Sampling rate (KHz) : %d" % sr2)                  
print("Audio length (seconds) : %.2f" % (len(y2) / sr2)) 
```

```python
# Use a pre-computed log-power Mel spectrogram
S = librosa.feature.melspectrogram(y=y3, sr=sr3, n_mels=128, fmax=8000)
mfccs3 = librosa.feature.mfcc(S=librosa.power_to_db(S))
print("y : ", y3)
print("sr : ", sr3)
print("y길이 : ", len(y3))
print("Sampling rate (KHz) : %d" % sr3)                  
print("Audio length (seconds) : %.2f" % (len(y3) / sr3)) 
```

```python
import matplotlib.pyplot as plt
import librosa.display
fig, ax = plt.subplots()
img = librosa.display.specshow(mfccs1, x_axis='time', ax=ax)
fig.colorbar(img, ax=ax)
ax.set(title='MFCC1')
```

```python
import matplotlib.pyplot as plt
import librosa.display
fig, ax = plt.subplots()
img = librosa.display.specshow(mfccs2, x_axis='time', ax=ax)
fig.colorbar(img, ax=ax)
ax.set(title='MFCC2')
```

```python
import matplotlib.pyplot as plt
import librosa.display
fig, ax = plt.subplots()
img = librosa.display.specshow(mfccs3, x_axis='time', ax=ax)
fig.colorbar(img, ax=ax)
ax.set(title='MFCC3')
```
