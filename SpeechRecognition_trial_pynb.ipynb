```
# .wav 파일에서 텍스트 추출하기

import speech_recognition as sr

recognizer = sr.Recognizer()
recognizer.energy_threshold = 300

## wav 파일 읽어오기
downy_audio = sr.AudioFile("/content/downy.wav")

with downy_audio as source:
    downy = recognizer.record(source)

recognizer.recognize_google(audio_data=downy, language="ko-KR")
```

```
from pydub import AudioSegment

downy_wav = AudioSegment.from_file("/content/downy.wav")

print(f"원채널수: {downy_wav.channels}")

downy_channels = downy_wav.split_to_mono()
print(f"분리한 채널: \n 남자: {downy_channels[0].channels}, \n 여자: {downy_channels[1].channels}")

# Save new channels separately
downy_channel_sohn = downy_channels[0]
downy_channel_kim = downy_channels[1]
```

```
def descriptive_stats(file_name):

    audio_segment = AudioSegment.from_file(file_name)
    
    print(f"- 채널수: {audio_segment.channels}")
    print(f"- 샘플 길이: {audio_segment.sample_width}")
    print(f"- 프레임율(frame rate): {audio_segment.frame_rate}")
    print(f"- 프레임폭(frame width): {audio_segment.frame_width}")
    print(f"- 길이 (ms): {len(audio_segment)}")
    
    return audio_segment

# Try the function
descriptive_stats('/content/downy.wav')
```
