```python
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

# Loading audio files with Librosa
출처 : https://www.youtube.com/watch?v=WJI-17MNpdE

```python
file = '../../nmb_data/teamvoice/testvoice_F2.wav'
```

```python
# load audio files with librosa
signal, sr = librosa.load(file)
print(signal.shape)
```

# Extracting MFCCs

```python
mfccs = librosa.feature.mfcc(y=signal, n_mfcc=13, sr=sr)
print(mfccs.shape)
```

# Visualising MFCCs

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

x : time    
y : coeeficient    

# Computing first / second MFCCs derivatives

```python
# calculate delta and delta2 MFCCs

delta_mfccs = librosa.feature.delta(mfccs)
delta2_mfccs = librosa.feature.delta(mfccs, order=2)

print(delta_mfccs.shape)
print(delta2_mfccs.shape)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta_mfccs, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(delta2_mfccs, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```

```python
mfccs_features = np.concatenate((mfccs, delta_mfccs, delta2_mfccs))
print(mfccs_features.shape)
```

```python
plt.figure(figsize=(25, 10))
librosa.display.specshow(mfccs_features, 
                         x_axis="time", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.show()
```
